```python
import pandas as pd
import numpy as np
```

```python
games_url = 'https://andybek.com/pandas-games'
```

```python
games = pd.read_csv(games_url)
```

```python
games.info()
```

```python
games.describe()
```

```python
#Q1 What are Sales across regions
```

```python
 games.loc[:,['NA_Sales',	'EU_Sales',	'JP_Sales',	'Other_Sales',	'Global_Sales']].max(axis=0)
```

```python
sales = games.loc[:, ['Platform', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']]
```

```python
sales.head()
```

```python
#Q2 What are Sales across each Platform
```

```python
sales.loc[games.Platform == 'X360'].sum(numeric_only=True)
```

```python
sales.loc[games.Platform == 'PS3'].sum(numeric_only=True)
```

```python
#Q3 What are Sales across each Platform
```

```python
sales.groupby('Platform').sum()
```

```python
#Q4 What are Sales across each Platform using Series
```

```python
sales_ser = games.loc[:, ['Platform', 'Other_Sales']].set_index('Platform').squeeze()
type(sales_ser)
```

```python
sales_ser.groupby('Platform').sum().sort_values(ascending=False)
```

```python
#Q4 Which Platform has Sales > 50 using Series
```

```python
def more_50(df):
  return games['Other_Sales'].sum()>50
```

```python
sales_ser
```

```python
sales_ser.groupby('Platform').filter(more_50)
```
