```python
import pandas as pd
import numpy as np
```

```python
#Read the files
data = pd.read_excel('/content/Cost of car insurance by state.xlsx')
hexmap = pd.read_excel('/content/hexmap.xlsx')
```

```python
data.columns = data.iloc[0]
data = data[1:]
```

```python
new_data= data.loc[:,['State',	'Full coverage',	'Minimum coverage',	'Difference']]
new_data.head()
```

```python
hexmap.head()
```

```python
#Merge

merge = pd.merge(new_data, hexmap , how='inner')
merge
```

```python
#Merge with Left and Right Params
mergelr= pd.merge(new_data, hexmap , left_on='State', right_on='State').drop(columns=['Row', 'Column'], axis=1)
mergelr
```

```python
#Concat with  Inner Join

ijoin = pd.concat([new_data, hexmap], join='inner')
ijoin
```

```python
#Concat with  Outer Join

ojoin = pd.concat([new_data, hexmap], join='outer').drop(columns=['Row', 'Column'], axis=1)
ojoin
```
